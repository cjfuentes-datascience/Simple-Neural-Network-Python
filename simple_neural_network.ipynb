This is like opening the box of Legos and picking out the pieces we want. "numpy" helps us with numbers, "Sequential" is the base for our robot, and "Dense" is a type of Lego block we'll use.



```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
```

Now, let's create our examples:


```python
X = np.array([[1], [2], [3], [7], [8], [9]])  # Input numbers
Y = np.array([[0], [0], [0], [1], [1], [1]])  # Whether they are small (0) or big (1)
```

Let's start building our robot! We'll use the "Sequential" base:


```python
model = Sequential()
```

We're adding a "Dense" block with 1 output (the answer), 1 input (the number), and "sigmoid" activation, which helps the robot decide if the answer is 0 or 1.



```python
model.add(Dense(1, input_dim=1, activation='sigmoid'))
```

We tell our robot to learn using "binary_crossentropy" (a way to check if it's right or wrong) and "adam" (a way to improve its answers). We also want to see how accurate it is.



```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

We give our robot the "X" and "Y" examples and tell it to learn from them 1000 times (epochs). "verbose=0" means it won't show us messages while learning.



```python
model.fit(X, Y, epochs=1000)
```

Test our robot!

We give our robot new numbers (4 and 10) and ask it if they are small or big. It will give us an answer between 0 and 1. The closer to 0, the smaller, and the closer to 1, the bigger.



```python
test_numbers = np.array([[4], [10]])
predictions = model.predict(test_numbers)
predictions
```
